```python
import numpy as np
#Used for array manipulation
import matplotlib.pyplot as plt
#pyplot will be used to plot the graphs and required plots
```

```python
data = np.loadtxt('data_Seminar-Supervised-Classification-K-NN-Performance-Metrics.txt')
#Load the text file which contains the required data
```

```python
X = data[:,0:2]
#X will contain the first 2 columns 
y = data[:,2]
#y will contain the last column of the text file 
plt.figure()
plt.scatter(X[:,0],X[:,1],c=y,marker='*')
#X[:,0] will take the first column and X[:,1] will take the last column of X 
#in the scatter we have 2 different colors where yellow represents 1 and purple represents 0
plt.show()
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
#Import the train and test split module to split the data between training and testing
trainX,testX,trainy,testy = train_test_split(X,y,test_size=0.2,random_state=2)
#Train and test split the data, where test size is 20% of the whole dataset, where the random state will control the shuffling applied to the data before splitting
```

```python
from sklearn.neighbors import KNeighborsClassifier
#Import the KNN classifer
model = KNeighborsClassifier(n_neighbors=3)
#Set the number of neighbours to 3
model.fit(trainX,trainy)
#Fit the training data according to chosen 3 neighbours
```

```python
model7 = KNeighborsClassifier(n_neighbors=7)
#Set the number of neighbours to 7
model7.fit(trainX,trainy)
#Fit the training data according to chosen 7 neighbours
```

```python
from sklearn.metrics import confusion_matrix
#import metrics module which will be used to create a confusion matrix of the given data
cnf_matrix = confusion_matrix(testy,model.predict(testX))
#create the matrix

TP = cnf_matrix[0][0]
#Location of the True Positive Rate in the matrix
FP = cnf_matrix[0][1]
#Location of the False Positive Rate in the matrix
FN = cnf_matrix[1][0]
#Location of the False Negative Rate in the matrix
TN = cnf_matrix[1][1]
#Location of the True Negative Rate in the matrix
print('TP = ',TP)
print('FP = ',FP)
print('FN = ',FN)
print('TN = ',TN)

print('\nPerformance Metrics')
print('TPR = Sensitivity =  '+'{:.2f}'.format(TP/(TP+FN)))
#Calculate the Sensitivity of the matrix
print('TNR = Specificity =  '+'{:.2f}'.format(TN/(TN+FP)))
#Calculate the Specificity of the matrix
print('PPV = Precision = '+'{:.2f}'.format(TP/(TP+FP)))
#Calculate the Precision of the matrix
print('ACC = Accuracy = '+'{:.2f}'.format((TP+TN)/(TP+FP+FN+TN)))
#Calculate the Accuracy of the matrix
```

```python
from sklearn.metrics import f1_score
#Import the F1_score module to calculate the F1_score
print('f1_score = '+'{:.2f}'.format(f1_score(testy,model.predict(testX),pos_label=1)))
#Print the F1 score when the KNN is 3
```

```python
from sklearn.metrics import fbeta_score
#Import the Fbeta Score this can be used to calculate the Fscore for F1 and Fn but in this case I will be calculating F2 Scores
print('f2_score = '+'{:.2f}'.format(fbeta_score(testy,model.predict(testX),beta=2)))
#Print the F2 Scores when the KNN is 3
```

```python
def plot_confusion_matrix(cm,classes, normalize=False, title='Confusion matrix',cmap=plt.cm.Blues):
    plt.imshow(cm,interpolation='nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks,classes,rotation=45)
    plt.yticks(tick_marks,classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max()/ 2.
    import itertools
    for i,j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
        plt.text(j, i, format(cm[i,j],fmt),
                horizontalalignment='center',
                color = 'white' if cm[i,j] > thresh else 'black')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    #The function has been created to visualise the condusion matrix

                
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
#Import both the ROC and AUC score metrics and the curve 

probs = model.predict_proba(testX)
#Return the probability estimates for testX where KNN is 3

probs = probs[:,1]
fpr,tpr,thresholds = roc_curve(testy,probs)
#Build the curve where it uses the testy for y and probs for X. fpr(False Positive Rate), tpr (True Positive Rate) and thresholds

auc = roc_auc_score(testy,probs)
#Calculate the ROC and AUC based on testy and probs
print('\nArea Under the curve: AUC = %.3f/n' %auc)
#Print the area under the curve metrics

plt.figure()
#Plot a new figure
lw = 3
#Length width
plt.plot(fpr,tpr, color='darkorange',lw=lw, label='ROC curve(area = %0.2f)' %auc)
#Plot the roc curve
plt.plot([0,1], [0,1], color='navy',lw=lw,linestyle='--')
plt.xlim([0.0,1.0])
#Plot the X limits
plt.ylim([0.0,1.05])
#Plot the y limits
plt.xlabel('False Positive Rate')
#Plot the x label
plt.ylabel('True Positive Rate')
#Plot the y label
plt.title('Reciver Operating Characteristics')
#Plot the Title of the diagram
plt.legend(loc='lower right')
#Plot the legend at the bottom right of the screen
plt.show()
```
