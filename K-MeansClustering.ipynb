```python
import matplotlib.pyplot as plt
#pyplot will be used to plot the data points
from sklearn.cluster import KMeans
#from the sklearn cluster we will import Kmeans module to create clusters through Kmeans
import numpy as np 
#Used for array manipulation
#Import the required modules 
```

```python
plt.rcParams['figure.figsize'] = (16,9)
#The code above will create a window of 16 width, 9 length
X = np.loadtxt("data_Kclusters.txt")
#Load the data from the text file
plt.scatter(X[:, 0], X[:, 1],)
#Create a scatter graph where values in X[:,0] is the x values and y values in X[:,1]
plt.show()
#Output the scatter graph

```

```python
WCSS  = []
#An array  which will store the results for the WCSS which will be used for the plotting
K = range(1,10) #range between 1 and 10
for this_k in K:
    kmeanModel = KMeans(n_clusters = this_k) #use the kmeans function where the cluster is the value at a given range, stored in variable kMeanmodel
    kmeanModel.fit(X)#Fit the model based on the data 
    WCSS.append(kmeanModel.inertia_)#Inertia, whcih measures how well the data is clusters will be appended into the WCSS array.

plt.figure()#Plot the new figure
plt.plot(K,WCSS,'bx-')#plt K onto the x axis and WCSS onto the y axis, where the colour is blue, x is the marker and - is the line type
plt.xlabel('k')#the label for x as K
plt.ylabel('WCSS')# the label for y as WCSS
plt.title('The Elbow method showing the optimal number of k')#The title for the plot
for i, txt in enumerate(WCSS):
    plt.annotate(round(txt,2),(K[i],WCSS[i]))#In the plot round the text to 2 dp, annonating the k and WCSS points in the plot 

```

```python
K_elbow = 3 #Set the number of k based on the optimum value as found above
kmeans = KMeans(n_clusters = K_elbow)#Use the k means function where we have set the optimal value
kmeans = kmeans.fit(X)#Fit the kmeans based on the data provided
labels = kmeans.predict(X) #Use the predict method to predict the clusters
C = kmeans.cluster_centers_#C shows the cluster centers

plt.figure()#Plot the new figure
plt.scatter(X[:, 0], X[:, 1], c=labels)#Plot the data points
plt.scatter(C[:, 0], C[:, 1],  marker='.', c='r', s=1000)#plot the intial centroids
plt.show() #Show the figure
```
